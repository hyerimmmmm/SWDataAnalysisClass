### 동영상 출력하기
- cv2.waitKey(밀리초): 동영상이 실행되는 시간에 대해서 제어
- cv2.waitKey() : 사용자의 키보드로부터 입력된 아스키 코드값을 반환 -> 윈도우창 꺼지는 기능

```python
cap = cv2.VideoCapture('./data/dogs.mp4') # 영상(비디오 객체) 가져옴
while True: # 계속 실행, 동영상 : 연속적인 이미지
    ch, frame = cap.read() # ch : 성공 여부(T or F), frame : 받아온 이미지

    # 실패할 경우 while 멈추기
    # False일 경우 실행되는 문장
    if not ch:
        print('동영상 재생 끝 / 읽기 실패')
        break

    # 제대로 읽었을 경우
    cv2.imshow('video1',frame)

    # q 클릭 시 동영상 윈도우 창 꺼지도록 설정
    key = cv2.waitKey(33) # 속도 제어, 사용자 입력값은 key에 저장
    if key == ord('q'):
        print('사용자에 의해 영상 종료')
        break

cap.release() # 카메라 연결 자원 해제
cv2.destroyAllWindows()
    
    
```

```python
cap = cv2.VideoCapture(0) # 영상(비디오 객체) 가져옴
cnt = 0
while True: # 계속 실행, 동영상 : 연속적인 이미지
    ch, frame = cap.read() # ch : 성공 여부(T or F), frame : 받아온 이미지

    # 실패할 경우 while 멈추기
    # False일 경우 실행되는 문장
    if not ch:
        print('동영상 재생 끝 / 읽기 실패')
        break

    # 제대로 읽었을 경우
    cv2.imshow('video1',frame)

    # q 클릭 시 동영상 윈도우 창 꺼지도록 설정
    key = cv2.waitKey(33) # 속도 제어, 사용자 입력값은 key에 저장
    if key == ord('q'):
        print('사용자에 의해 영상 종료')
        break

    # 사진 찍기, 프레임 저장하기
    if key == ord('w'):
        cnt += 1
        print(f'./data/dog_img{cnt}.png 사진 저장')
        cv2.imwrite(f'./data/dog_img{cnt}.png', frame, params=[cv2.IMWRITE_PNG_COMPRESSION, 0])
        # imwrite 이미지 저장 함수
        # imwrite(경로, 저장할 이미지, 이미지 저장 시의 옵션(압축정도, 0일경우 최고화질))


cap.release() # 카메라 연결 자원 해제
cv2.destroyAllWindows()
    
    
```

```python
plt.xticks([]) # 격자 표시 제거
plt.yticks([]) # 격자 표시 제거
for i in range(1,5):
    selca = cv2.imread(f'./data/dog_img{i}.png')
    selca = cv2.cvtColor(selca, cv2.COLOR_BGR2RGB)
    plt.subplot(2,2,i)
    plt.xticks([]) # 격자 표시 제거
    plt.yticks([]) # 격자 표시 제거
    plt.imshow(selca)
```
