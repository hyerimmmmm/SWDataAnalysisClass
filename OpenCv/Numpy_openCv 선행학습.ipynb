### Numpy 데이터 접근

- 불리언 인덱싱: 논리 연산자를 활용하여 True, False (True)결과값에 접근

- 조건에 맞는 데이터 접근하기(불리언 인덱싱)

```python
sample_arr = [True, False] # 불리언값 추출용

np.random.seed(1) # 난수 생성기 시드 고정

bool_arr = np.random.choice(sample_arr, size = 10)
display(bool_arr)

tmp_data = np.arange(10) # 0~9 배열 생성
display(tmp_data)

display(tmp_data[bool_arr]) # tmp_data 배열에서 bool_arr가 True인 경우만
```

```python
# print, display 차이
print(bool_arr)
display(bool_arr)
```

```python
# 성적데이터 생성 후 조건에 맞는 데이터 접근
# 8개의 성적, 8명의 학생 번호
# 40~100 랜덤 숫자값 8개 배열 표현되게 생성

np.random.seed(7) # 랜덤 시드 고정
score_arr = np.random.randint(40, 101, size = 8) # 랜덤으로 40~100 사이 숫자 8개 생성

display(score_arr)

stu_num = np.arange(1, 9) # 각 학생 번호

display(stu_num)
```

```python
# 논리연산 == != > < >= <=
# 75점 이상인 성적은 몇개 있을까?
len(score_arr[score_arr >= 75])
score_arr[score_arr >= 75].size
```

```python
# 75점 이상인 성적을 가진 학생 번호
stu_num[score_arr >= 75]

# 데이터 분석 쪽에서 자주 사용
# 이미지 분야 numpy 인덱싱, 슬라이싱 기법 데이터 접근
```
