### 산술연산자

```python
num1 = 10
num2 = 7
print(num1 / num2) # 나눈값
print(round(num1 / num2, 2)) # 나눈값을 반올림
print(num1 // num2) # 몫
print(num1 % num2) # 나머지
```

### 비교연산자
- 불리언 값을 반환(true or false)

```python
# 조건식 사용

3 > 5
```

```python
a = 3
b = 7

print(a > b)
print(a < b)
print(a >= b)
print(a <= b)

print(a == b)
print(a != b)

print()

# 논리연산자 not은 논리값을 뒤집음
print(not a > b)
print(not a < b)
print(not a >= b)
print(not a <= b)

print(not a == b)
print(not a != b)
```

### 논리연산자
- and: 값이 모두 true 일 경우
- or: 값 중에 하나라고 true 일 경우

### 멤버연산자 in

```python
s = '콩국수에 들어갈 건 소금 뿐'

# 조건식 입력
# 소금이라는 문자가 s 안에 있는가?
'소금' in s
```

```python
s = '콩국수에 들어갈 건 소금 뿐'

# 조건식 입력
# 소금이라는 문자가 s 안에 있는가?
# 위의 질문에 대한 결과값을 not을 사용해 뒤집음
'소금' not in s
```

### 삼항 연산자
- 조건문의 단순화
- a if 조건식 else b
- a = 참일 때 실행할 내용
- b = 거짓일 때 실행할 내용

```python
# 삼항연산자

java_score = int(input('자바 점수 입력: '))

# 만약에 자바 점수가 60점 이상이라면 합격, 그렇지 않다면 불합격
'합격' if java_score >= 60 else '불합격'
```

```python
num = int(input('정수 입력: '))

f'{num}는(은) 홀수입니다.' if num % 2 != 0 else f'{num}는(은) 짝수입니다.'
```
