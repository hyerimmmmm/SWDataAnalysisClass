### 함수의 목적
- 반복적인 프로그래밍을 피할 수 있다
- 모듈화로 인해 가독성이 좋아진다
- 프로그램에 문제가 발생하거나 기능의 변경이 필요할 때 손쉽게 유지보수가 가능하다

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))

def number_sub(num1, num2):
    result = num1 - num2
    return result
    
result = number_sub(num1, num2)
print(result)
```

```python
# 문자열끼리의 덧셈

str1 = ''
str1 += '설'
str1 += '탕'

str1
```

```python
# 방법 1
s = input("문자열 입력 >> ")

def s_replace(str):
    str_result = ''
    for i in str:
        if i != 'ㅋ':
            str_result += i
    return str_result
            
result = s_replace(s)
result
```

```python
# 방법 2
# replace()
# 문자열.replace(바뀔 문자, 바꿀 문자)

str1 = '콩국수에는 설탕을 넣어야 제맛이다'
print('바꾸기 전: ', str1)

print('바꾼 후: ', str1.replace('설탕', '소금'))

def s_replace2(str):
    return str.replace('ㅋ', '')
result = s_replace2(s)
result
```

```python
def divisor(num1):
    '''
    약수를 구해서 출력해주는 함수
    여러줄 입력 가능
    '''
    for i in range(1, num1 + 1):
        if num1 % i == 0:
            print(i, end = ' ')
```

```python
divisor(10)
```
