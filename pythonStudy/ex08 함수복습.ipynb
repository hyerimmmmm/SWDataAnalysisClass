```python
num1 = int(input('    >> '))
num2 = int(input('    >> '))

def number_sub(n1, n2):
    return n1 - n2

result = number_sub(num1, num2)
print(result)
```

```python
s = input('  >> ')

def s_replace(str):
    result = ''
    for i in str:
        if i != '':
            result += i
    return result
        
result = s_replace(s)
result
```

```python
num1 = int(input('    >> '))
num2 = int(input('    >> '))
op = input(' (+, -) >> ')

def cal(n1, n2, str1):
    if str1 == '+':
        return n1 + n2
    elif str == '-':
        return n1 - n2
    else:
        return '  '

result = cal(num1, num2, op)
print(' : ', result)
```

```python
import pandas as pd
```

```python
pop = pd.Series([9600000, 3340000, 1480000, 2410000],
                index = ['', '', '', ''])

pop
```

```python
#   100 
pop + 100
```

```python
#  100   
def plus_100(x):
    return x + 100
```

```python
plus_100(5)
```

```python
# apply  
# Series   
pop.apply(plus_100)
```

```python
#   pop 
# pop   return  pop + 100 
plus_100(pop)
```

```python
# 
# 1. plus_100(pop): pop   ()  -> : pop + 100
# 2. pop.apply(plus_100):   pop    -> 9600000 +100, 3340000 + 100, ...
```

```python
plus_100(pop.apply(''))
```

```python
# 300    (  0 )

def con_return(x):
    if x >= 3000000:
        return x
    else:
        return 0
```

```python
con_return(350000)
```

```python
# apply  
pop.apply(con_return)
```

```python
# apply 
#   pop  
con_return(pop)
```
