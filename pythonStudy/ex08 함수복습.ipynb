```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))

def number_sub(n1, n2):
    return n1 - n2

result = number_sub(num1, num2)
print(result)
```

```python
s = input('문자열 입력 >> ')

def s_replace(str):
    result = ''
    for i in str:
        if i != 'ㅋ':
            result += i
    return result
        
result = s_replace(s)
result
```

### apply 함수
- 각 요소에 특정 함수를 적용시켜주는 함수

```python
import pandas as pd
```

```python
pop = pd.Series([9600000, 3340000, 1480000, 2410000],
                index = ['서울', '부산', '광주', '대구'])

pop
```

```python
# 각 값에 100을 더하기
pop + 100
```

```python
# 값을 100씩 더해주는 함수 정의
def plus_100(x):
    return x + 100
```

```python
plus_100(5)
```

```python
# apply 함수 사용
# Series의 개별적 요소에 적용
pop.apply(plus_100)
```

```python
# 함수에 인수로 pop를 넣어보자
# pop가 함수에 들어가서 return 값으로 pop + 100 반환
plus_100(pop)
```

```python
# 차이점
# 1. plus_100(pop): pop 자체를 함수의 인수로서(입력값) 전달 -> 반환값: pop + 100
# 2. pop.apply(plus_100): 함수가 각각 pop의 요소에 개별적으로 접근 -> 9600000 +100, 3340000 + 100, ...
```

```python
plus_100(pop.apply('서울'))
```

```python
# 300만 이상인 값만 표시해주는 함수(그 외에는 0 반환)

def con_return(x):
    if x >= 3000000:
        return x
    else:
        return 0
```

```python
con_return(350000)
```

```python
# apply 함수 사용
pop.apply(con_return)
```

```python
# apply 미사용
# 함수 안에 pop 인수로 대입
con_return(pop)
```

```python
# Lambda x: 익명 합수 정의(잠깐 사용할 함수 정의)
# apply 함수과 자주 함께 사용
pop.apply(lambda x: x + 200)
```

```python
# 숫자에 천 단위 구분 기호 추가
pop.apply(lambda x: f'{x: ,}')
# 문자열 포매팅 f'{x: }: x 라는 변수에 특정 포매팅을 적용시키고 싶을 때 사용(, 는 천단위 구분하라는 기호)
```
