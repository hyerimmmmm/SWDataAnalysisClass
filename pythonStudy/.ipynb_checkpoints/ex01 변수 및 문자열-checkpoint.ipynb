### 문자열 포매팅
- 문자열 안의 특정한 값을 바꿔야 하는 경우
- 문자열 안에 변수를 넣고 싶을 때 사용

```python
s1 = '오늘은 3월 25일 입니다.'

s2 = '오늘은 3월 26일 입니다.'

print(s1)

# %기호를 사용하는 방법

day = 22

# 정수형 변수를 넣을 자리에 %d, 마지막에 %들어갈 변수명
# %d: 정수형
# %f: 실수형
# %s: 문자형

s1 = '오늘은 3월 %d일 입니다.'%day

print(s1)
```

```python
# day = day + 1
day += 1

print('오늘은 3월 %d일 입니다'%day)
```

```python
# format 함수를 사용한 포매팅
# .format(첫번째 변수, 두번째 변수, ...)
# 자료형이 달라져도 상관없음

month = 3
day = '이십오'

s = '오늘은 {}월 {}일 입니다.'.format(month, day)

print(s)
```

```python
# f문자열을 사용한 포매팅
# 현재 가장 많이 쓰이는 방법, python3 부터 제공하는 기능

month = 4
day = 5

s = f'오늘은 {month}월 {day}일입니다.'

print(s)
```

```python
x = 100
y = 200
sum2 = x + y
print(f'{x}와 {y}의 합은 {sum2}입니다.')
```

### 실습. 팀장의 이름을 입력받아서 팀장의 성과 이름을 출력해보자
- 입력된 값이 저장된 name에서 인덱싱과 슬라이싱
- 문자열 포매팅

```python
name2 = input('팀장의 이름을 입력해주세요: ')
firstName = name2[0]
lastName = name2[1:]

print(f'팀장의 성은 {firstName}이고 이름은 {lastName}입니다.')
```
