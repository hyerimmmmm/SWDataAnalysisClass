### 반복문
- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문
  - while: 반복횟수가 명확하지 않을 때 사용(반복횟수를 임의로 조절하는 것은 가능)
  - for: 반복횟수가 명확할 때 사용
  - break: 반복문을 나가는 기능

#### 파이썬 최고를 5번 출력하세요

```python
cnt = 0 # 반복횟수를 조절하기 위한 변수
while cnt < 5:
    print('파이썬 최고')
    cnt += 1
```

```python
n = 1
while True:
    if n > 5:
        break
        
    print('파이썬 최고')
    n += 1
```

```python
import random

num = random.randint(1, 50) # 1부터 50까지 숫자중 랜덤

while True:
    num_1 = int(input('숫자를 입력하세요 >> '))
    if num_1 == num:
        print('정답을 맞추셨습니다.')
        break
    elif num_1 > num:
        print(num_1, '보다 작습니다.')
    else:
        print(num_1, '보다 큽니다.')
```

```python
while True:
    num1 = int(input('첫번째 정수 입력 >> '))
    num2 = int(input('두번째 정수 입력 >> '))
    print('두 정수의 합 : ', num1 + num2)
    if num1 == 0 and num2 == 0:
        print('프로그램이 종료되었습니다.')
        break
```

### range() 함수
- 필요한 만큼의 숫자를 만들어내는 함수
- range(시작할 숫자, 종료할 숫자(끝값은 포함하지 않음), 증가량)

```python
range(1, 10, 1) # 1부터 1씩 증가하며 9까지 생성
```

```python
range(10, 1, -1) # 10부터 1씩 감소하며 2까지 생성
```

```python
for i in range(10, 1, -1):
    print(i, end = ' ')
```

```python
for i in range(10, 1, -1):
    print('파이썬 최고')
```

```python
for i in range(9):
    print('파이썬 최고')
```

```python
# range(0, n, 1)
# 증감량이 1일 경우 생략이 가능

# -> range(0, n)
# 증감량이 1이면서 시작 숫자가 0이면 시작 숫자 생략 가능

# -> range(n)
```

```python
num = int(input('정수 입력 >> '))
print(num,'의 약수: ', end = '')

for i in range(1, num + 1):
    if num % i == 0:
        print(i, end = ' ')
```

### 구구단을 만들어보자
- 이중 for 문

```python
for i in range(1, 10):
    print(f'2 x {i} = {2 * i}')
```

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(f'{i} x {j} = {i * j}')    
    print()
```

```python
# 학생별 점수
list2 = [40, 80, 70, 90, 50] # 1-5번 학생의 점수
n = 1 # 학생 번호를 담아 줄 변수

for i in list2:
    print(f'{n}번 학생 점수는 {i}점 입니다.')
    n += 1
```

```python
score_list = [90, 45, 70, 60, 55]
j = 1

for i in score_list:
    if i >= 60:
        print(f'{j}번 학생은 합격입니다.')

    else:
        print(f'{j}번 학생은 불합격입니다.')
    j += 1
```

### 종합 실습
- list, if, for, append(), range(), len() 사용

```python
list4 = [1, 2, 3, 4, 5, 6, 7]
list5 = [3, 4, 5, 6]
list6 = []

num = 0
max = 0

if len(list4) > len(list5) : 
    num = len(list5)
else:
    num = len(list4)

if len(list4) > len(list5):
    max = len(list4)
else:
    max = len(list5)


for i in range(num):
    answer = list4[i] + list5[i]
    list6.append(answer)

for i in range(num, max):
    if len(list4) > len(list5):
        list6.append(list4[i])
    else:
        list6.append(list5[i])
        
print(list6)
```

```python
list4 = [1,2,3,4,5,6,7]
list5 = [3,4,5,6]
list6 =[]

print('list4 :', list4)
print('list5 :', list5)

# 몇번 append?
# 더 짧은 리스트의 길이만큼 더한값이 넣어짐
# n

# 두 리스트의 길이의 차이만큼, 더 긴 리스트의 값이 들어간다.
# n2

if len(list4) < len(list5):
    n = len(list4)
    n2 = len(list5) - len(list4)
else:
    n = len(list5) # 4
    n2 = len(list4) - len(list5)  # 7-4 = 3

# 앞에 더해지는 부분 append()
for i in range(n):
    list6.append(list4[i]+list5[i])

print('중간점검 :', list6)

# 리스트의 뒤부분은 한쪽(더 긴 리스트)에만 있는 값 append()
for i in range(n2): # range(3)
    if len(list4)<len(list5):
        list6.append(list5[n+i])
    else: # list 4가 더 길다면~
        list6.append(list4[n+i]) # list4[4], list4[5], list4[6] // 4+0, 4+1, 4+2

print('최종 :', list6)
```
