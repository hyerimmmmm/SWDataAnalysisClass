### 리스트
- 순서가 있는 (인덱스 번호가 존재하는) 수정 가능한 객체의 집합
- 대괄호 '[ ]'로 작성되어지며, 리스트 내부의 요소들은 콤마 ','로 구분

### 리스트의 인덱싱

```python
list1 = [2, 5, 7, 9, 10]
```

```python
list1
```

```python
print(list1)

# 리스트의 0번 요소 출력
print(list1[0])

# 리스트의 3번 요소 출력
print(list1[3])

# 리스트의 -1번 요소 출력
print(list1[-1])
```

### 리스트의 슬라이싱
- 인덱싱의 경우 인덱싱 할 때 리스트 안의 요소를 그대로 가져옴
- 슬라이싱의 경우, 요소들을 리스트에 담아 리스트의 형태로 가져옴

```python
list2 = [0, 1, 2, 3, 4, 5]
```

```python
# 1, 2 가져오기
list2[1:3]
```

```python
# 0, 1
list2[: 2]
```

```python
# 3
list2[3]
```

```python
list2[3:4]
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
```

```python
print('첫 번째 요소의 두 번째 요소: ', matrix[0][1])
print('두 번째 요소 전체: ', matrix[1])
print('두 번째 요소부터 끝까지: ', matrix[1:])
print('두 번째 요소에서, 첫번째부터 두번째 요소까지: ', matrix[1][:2])
```

```python
# list의 요소 변경 및 추가

list5 = [0, 1, 2]
print('변경 전', list5)

# 요소 변경
list5[2] = 65
print('변경 후', list5)
```

```python
# list의 요소 추가
# 중요
# append: list 자료형에 포함된 메서드

list5.append(25);
print(list5)
```

```python
# 슬라이싱으로 list의 요소 변경

list5 = [0, 1, 2, 3, 4, 5, 6]
print('변경 전: ', list5)

list5[2:5] = [4, 8, 2]
print('변경 후: ', list5)

list5[2:5] = [482]
print('변경 후: ', list5)

list5[2:5] = ['a', 'b', 'c', 99, 6, 7]
print('변경 후: ', list5)
```

```python
array = [1, 2, 3, 4, 5]

array[4] = 7
print(array)
```

```python
# 인덱싱으로 값을 변경하는 경우 해당 요소에 요소 자체를 그대로 변경
array[1] = ['a', 'b', 'c']
print(array)
```

```python
# 슬라이싱으로 값을 변경하는 경우 변경할 값(리스트)에 담긴 값들의 각각의 요소로 변경

array[3:] = ['d', 'e', 'f', 'g']
print(array)
```

```python
fruits = ['사과', '포도', ['수박', '멜론'], '오렌지']
```

```python
apple = fruits[0]
melon = fruits[2][1]
orange = fruits[3]

print(apple, melon, orange)
```

```python
choice_list = []
choice_list.append(apple)
choice_list.append(melon)
choice_list.append(orange)

print(choice_list)
```

```python
# 한번에 list에 값 넣기
choice_list2 = []

# extend: list 안에 여러개의 값을 한번에 추가
# list의 형태로 값을 추가해야 함
choice_list2.extend([apple, melon, orange])

choice_list2
```

### list에 사용되는 다양한 함수들

```python
lst = [1, 2, 3, 4, 5, 6 ,7, 8, 9, 0, 5]

# len: list의 길이(요소의 개수)를 타나내는 함수
# len(리스트명)
# 내장 함수라 list 한정은 아님(문자열 등에서도 사용가능)
print(len(lst))

str = 'asdfadfasfdsf'
print(len(str))
```

```python
# index: 특정 값의 인덱스 반환
# 리스트명.index(찾고자하는 값)
# 중복되는 값이 있는 경우에는 그 중 제일 처음나오는 값의 위치만 알려줌
lst.index(6)
```

```python
# insert: list의 원하는 위치에 값을 추가하는 함수
# 리스트명.insert(원하는 인덱스 위치, 추가할 값)
print('추가 전: ', lst)
lst.insert(5, 34)
print('추가 후: ', lst)
```

```python
# list 값을 삭제하는 방법 1
# del 사용(여기저기 사용가능)
list7 = [33, 44, 55, 66, 77, 88, 99]
print('삭제 전: ', list7)

del list7[3]
print('삭제 후: ', list7)
```

```python
# list 값을 삭제하는 방법 2
# remove: list 값을 사용해 삭제하는 함수
# 리스트명.remove(삭제할 값)
list7.remove(44)
print('삭제 후: ', list7)
```

```python
# 2-1. 동일한 값이 있는 경우
# 앞에 있는 값만 삭제하고 종료
# 다른 셀에서 계속 삭제하면 더 이상 해당 값이 없어서 에러가 날 때까지 계속 삭제할 수 있음 -> 반복문 돌리기
list8 = [33, 44, 55, 99, 66, 77, 88, 99]
print('삭제 전: ', list8)

list8.remove(99)
print('삭제 후: ', list8)
```

```python
# list 값을 삭제하는 방법 3
# 슬라이싱을 활용해서 삭제(feat.del)
list7 = [33, 44, 55, 66, 77, 88, 99]
print('삭제 전: ', list7)

del list7[1:3]
print('삭제 후: ', list7)
```

```python
# sort: 오름차순 정렬해주는 함수
# 리스트명.sort()

list7 = [33, 44, 55, 77, 88, 99, 66, 234, 23, 232, 645, 223]
list7.sort()
print(list7)
```

```python
# reverse: 역순 정렬해주는 함수(내림차순 아님)
# 리스트명.reverse()

list7 = [33, 44, 55, 77, 88, 99, 66, 234, 23, 232, 645, 223]
list7.reverse()
print(list7)
```

```python
# 내림차순 정렬하는 방법
# 1. sort 실행   2. reverse 실행

list7 = [33, 44, 55, 77, 88, 99, 66, 234, 23, 232, 645, 223]
list7.sort()
list7.reverse()
print(list7)
```

```python
# in: 찾고자 하는 값이 안에 있으면 true 반환하는 연산자
list9 = ['a', 'b', 'c']

print('a' in list9)
print('d' in list9)
```
