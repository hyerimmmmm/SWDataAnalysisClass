## Matplotlib
### Mathdatas plot library
- 메타데이터를 도표나 그래프(시각화)로 출력시켜주는 라이브러리
- 메타데이터: 일정한 규칙에 따라 구조화, 표준화된 데이터
- 분석 결과를 수치만으로 설명하는 것보다 더 직관적으로 이해 가능
- 타인에게 분석 결과를 전달하기에 효과적

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 플롯 차트(plot chart): 선 그래프

```python
y = [2, 4, 5]

plt.plot(y) # 선 그래프를 가져오기
plt.show() # 콘솔에 차트를 출력
```

```python
# x, y 값 지정하여 선 그래프
x = [1, 2, 3, 4, 5]
y = [1, 2, 7, 8, 6]

plt.plot(x, y, ls = '--', lw = 1, c = 'gray', marker = 'P', mfc = 'black', mec = 'black', ms = 10, mew = '1')
# ls: 선 스타일
# lw: 선 굵기
# c: 선 색상
# marker: 마커의 종류
# mfc: 마커의 내부 색상
# mec: 마커의 선 색상
# ms: 마커의 크기
# mew: 마커의 선 굵기

plt.show()
```
