### 문제3.

```python
list1 = [1, 50, 27, 19, 37, 33, 5, 3, 48]
even = 0
odd = 0

for i in list1:
    if i % 2 == 0:
        even += 1
    else:
        odd += 1

print(f'짝수 {even}개, 홀수 {odd}개')
```

### 문제.4

```python
import numpy as np
```

```python
score_list = [['김미희', 89, 78, 99],
              ['서대희', 67, 88, 77],
              ['강예진', 100, 55, 96],
              ['차현석', 75, 100, 60]]
num = int(len(score_list))
python = 0
java = 0
db = 0

space1 = '     '
space2 = '      '

for i in range(num):
    python += score_list[i][1]
    java += score_list[i][2]
    db += score_list[i][3]

print('========================================================')
print('이름      Python    Java      DB        평균        학점')
print('========================================================')

for i in range(num):
    score_arr = np.array(score_list[i][1:])
    score_mean = int(score_arr.mean())
    if score_mean <= 100 and score_mean >= 0:
        if score_mean >= 90:
            grade = 'A'
        elif score_mean >= 80:
            grade = 'B'
        elif score_mean >= 70:
            grade = 'C'
        elif score_mean >= 60:
            grade = 'D'
        else:
            grade = 'F'
    print(score_list[i][0], '   ', score_arr[0], space1 if score_arr[0] == 100 else space2, score_arr[1], space1 if score_arr[1] == 100 else space2, score_arr[2], space1 if score_arr[2] == 100 else space2, score_mean, space2, grade)

print('========================================================')
print('평균      ', python // num, '      ', java // num, '      ', db // num)
```

### 문제5.

```python
import matplotlib.pyplot as plt
```

```python
X = [2, 3, 4, 5, 6, 7, 8]
Y1 = [5, 7, 3, 9, 1, 8, 6]
Y2 = [9, 5, 7, 2, 3, 7, 1]

plt.plot(X, Y1, label = 'Y1', c = 'red')
plt.plot(X, Y2, label = 'Y2', c = 'blue')

plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

plt.legend(loc = 'upper right', prop = {'size':10})

plt.show()
```

### 문제6.

```python
import pandas as pd
plt.rc('font', family = 'Pretendard')
```

```python
data_2017 = pd.read_csv('data/Traffic_Accident_2017.csv', encoding = 'euc-kr')
car_person = data_2017[data_2017['사고유형_대분류'] == '차대사람']
noa = car_person[car_person['주야'] == '야간']
```

```python
date = noa['요일'].value_counts()
date_sort = date[['월', '화', '수', '목', '금', '토', '일']]

date_sort
```

```python
y = date_sort.values
x = date_sort.index

plt.bar(x, y)

plt.ylim(0, 200)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고 건수')

plt.show()
```
