### 조건문
- 상황에 따라서 실행 흐름을 다르게 하는 문법
- 키워드: if, elif, else
- 콜론(:)과 들여쓰기(tap) 문법이 중요

```python
# 들여쓰기로 조건문의 실행 문장을 구분
if True:
    print('희진이는 팀장이다')
print('이 문장은 조건문 밖에 있는 문장이라 항상 실행된다')
```

```python
if False:
    print('설탕 국수는 존맛탱이다')
print('이 문장은 조건문 밖에 있는 문장이라 항상 실행된다')
```

```python
# 간단한 if 조건문
money = 9000

if money >= 10000:
    print('택시를 탄다')
else:
    print('버스를 탄다')
```

```python
num = int(input('정수 입력 >> '))

if (num % 3 == 0) and (num % 5 == 0):
    print('3과 5의 배수입니다.')
else:
    print('3과 5의 배수가 아닙니다.')
```

```python
num1 = int(input('첫 번째 정수 입력 >> '))
num2 = int(input('두 번째 정수 입력 >> '))

if num1 > num2:
    print('첫 번째 정수가 더 큽니다.')
elif num1 < num2:
    print('두 번째 정수가 더 큽니다.')
else:
    print('두 수가 똑같습니다.')
```

```python
score = int(input('점수 입력 >> '))

if 100 >= score >= 0:
    if score >= 90:
        print(f'{score}은(는) A학점입니다.')
    elif score >= 80:
        print(f'{score}은(는) B학점입니다.')
    elif score >= 70:
        print(f'{score}은(는) C학점입니다.')
    elif score >= 60:
        print(f'{score}은(는) D학점입니다.')
    else:
        print(f'{score}은(는) F학점입니다.')
else:
    print('잘못된 입력입니다.')
```

```python
score = int(input('점수 입력 >> '))
grade = 'A'

if 100 >= score >= 0:
    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    elif score >= 60:
        grade = 'D'
    else:
        grade = 'F'
    print(f'{score}점은 {grade}학점입니다.')
else:
    print('잘못된 입력입니다.')
```

```python
x = int(input('x좌표 >> '))
y = int(input('y좌표 >> '))

if x != 0 and y != 0:
    if x > 0 and y > 0:
        num = '제 1사분면'
    elif x < 0 and y > 0:
        num = '제 2사분면'
    elif x < 0 and y < 0:
        num = '제 3사분면'
    elif x > 0 and y < 0:
        num = '제 4사분면'
elif x == 0 and y == 0:
    num = '원점'
else:
    num = '잘못된 입력'
    

print(f'({x}, {y})는 {num}입니다.')
```
